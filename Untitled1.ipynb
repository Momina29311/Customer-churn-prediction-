```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Ali", "Sara", "Ahmed"]
})

df2 = pd.DataFrame({
    "EmpID": [1, 2, 4],
    "Salary": [50000, 45000, 60000]
})

# Merge
pd.merge(df1, df2, on="EmpID", how="inner")

pd.merge(df1, df2, on="EmpID", how="left")

pd.merge(df1, df2, on="EmpID", how="right")

# Concat
pd.concat([df1, df2], axis=0)
```
